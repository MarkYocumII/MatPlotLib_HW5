```python
#Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

#read csv data into dataframes
mouse_df = pd.read_csv(mouse_drug_data_to_load)
clinical_df = pd.read_csv(clinical_trial_data_to_load)

#merge data sets by Mouse ID
combined_mouse_df = pd.merge(mouse_df, clinical_df, how="outer", on="Mouse ID")
combined_mouse_df.head()
```

```python
#groupby drug and time point for mean tumor volumes

drug_groups = combined_mouse_df.groupby(["Drug", "Timepoint"])
tumors_df = pd.DataFrame()
tumors_df["Tumor Volume (mm3)"] = drug_groups["Tumor Volume (mm3)"].mean()

#reset index to get three columns where drug labels repeat
tumors_df = tumors_df.reset_index()
tumors_df.head()
```

```python
#determine standard errors for each measurement

#create new data frame that is exact copy of previous tumors_df
tumors_errors_df = tumors_df

#calculate standard error of tumor volumes from groupby object
tumors_errors_df = drug_groups["Tumor Volume (mm3)"].sem()

#reset index to get three columns where drug labels repeat
tumors_errors_df = tumors_errors_df.reset_index()
tumors_errors_df.head()
```

```python
#create new dataframes for tumor volumes and standard errors that are pivoted by timepoint and drug
volume_df = tumors_df.pivot(index="Timepoint", columns="Drug", values="Tumor Volume (mm3)")
errors_df = tumors_errors_df.pivot(index="Timepoint", columns="Drug", values="Tumor Volume (mm3)")

volume_df.head()
```

```python
errors_df.head()
```

```python
#Tumor volume plots for Capomulin, Infubinol, Ketapril, Placebo with error bars

#convernt dataframe index to values for x axis
x_axis = volume_df.index.values

#set of four errorbar line charts for each of the drugs studied
plt.errorbar(x_axis, volume_df["Capomulin"], errors_df["Capomulin"] ,color="red", marker = "o", linewidth=0.5)
plt.errorbar(x_axis, volume_df["Infubinol"], errors_df["Infubinol"] ,color="blue", marker = "^", linewidth=0.5)
plt.errorbar(x_axis, volume_df["Ketapril"], errors_df["Ketapril"] ,color="green", marker = "s", linewidth=0.5)
plt.errorbar(x_axis, volume_df["Placebo"], errors_df["Placebo"] ,color="black", marker = "D", linewidth=0.5)

#add grid, title, x- and y- labels, and legend to the chart

plt.grid()
plt.title("Tumor Response to Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.legend()
             
plt.savefig("Tumor_response.png")
    
plt.show()
```

```python
#store metastatic site data

#groupby drug and time point (initial group by object) for mean metastatic site

sites_df = pd.DataFrame()
sites_df["Metastatic Sites"] = drug_groups["Metastatic Sites"].mean()

#reset index to get three columns where drug labels repeat
sites_df = sites_df.reset_index()
sites_df.head()
```

```python
#store error bars for metastatic site data

#create new data frame that is exact copy of previous sites_df
sites_errors_df = sites_df

#calculate standard error of metastatic sites from groupby object
sites_errors_df = drug_groups["Metastatic Sites"].sem()

#reset index to get three columns where drug labels repeat
sites_errors_df = sites_errors_df.reset_index()
sites_errors_df.head()
```

```python
#pivoted tables for plotting

#create new dataframes for tumor volumes and standard errors that are pivoted by timepoint and drug
num_sites_df = sites_df.pivot(index="Timepoint", columns="Drug", values="Metastatic Sites")
errors_sites_df = sites_errors_df.pivot(index="Timepoint", columns="Drug", values="Metastatic Sites")

num_sites_df.head()
```

```python
errors_sites_df.head()
```

```python
#Metastatic site plots for Capomulin, Infubinol, Ketapril, Placebo with error bars

#convernt dataframe index to values for x axis
x_axis_sites = num_sites_df.index.values

#set of four errorbar line charts for each of the drugs studied
plt.errorbar(x_axis_sites, num_sites_df["Capomulin"], errors_sites_df["Capomulin"] ,color="red", marker = "o", linewidth=0.5)
plt.errorbar(x_axis_sites, num_sites_df["Infubinol"], errors_sites_df["Infubinol"] ,color="blue", marker = "^", linewidth=0.5)
plt.errorbar(x_axis_sites, num_sites_df["Ketapril"], errors_sites_df["Ketapril"] ,color="green", marker = "s", linewidth=0.5)
plt.errorbar(x_axis_sites, num_sites_df["Placebo"], errors_sites_df["Placebo"] ,color="black", marker = "D", linewidth=0.5)

#add grid, title, x- and y- labels, and legend to the chart

plt.grid()
plt.title("Metastatic Spread During Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Met. Sites")
plt.legend()

plt.savefig("Metastatic_Spread.png")
                
plt.show()
```

```python
#determine mouse counts (i.e. survival rates) for each drug

mouse_count_df = pd.DataFrame()
mouse_count_df["Mouse Count"] = drug_groups["Mouse ID"].count()
mouse_count_df = mouse_count_df.reset_index()
mouse_count_df.head()
```

```python
mouse_pivot_df = mouse_count_df.pivot(index="Timepoint", columns="Drug", values="Mouse Count")
mouse_pivot_df.head()
```

```python
#Metastatic site plots for Capomulin, Infubinol, Ketapril, Placebo with error bars

#convernt dataframe index to values for x axis
x_axis_mouse = mouse_pivot_df.index.values

#set of four errorbar line charts for each of the drugs studied
plt.plot(x_axis_mouse, mouse_pivot_df["Capomulin"]/mouse_pivot_df["Capomulin"].loc[0]*100, color="red", marker = "o", linewidth=0.5)
plt.plot(x_axis_mouse, mouse_pivot_df["Infubinol"]/mouse_pivot_df["Infubinol"].loc[0]*100, color="blue", marker = "^", linewidth=0.5)
plt.plot(x_axis_mouse, mouse_pivot_df["Ketapril"]/mouse_pivot_df["Ketapril"].loc[0]*100, color="green", marker = "s", linewidth=0.5)
plt.plot(x_axis_mouse, mouse_pivot_df["Placebo"]/mouse_pivot_df["Placebo"].loc[0]*100, color="black", marker = "D", linewidth=0.5)

#add grid, title, x- and y- labels, and legend to the chart

plt.grid()
plt.title("Survival During Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Survival Rate (%)")
plt.legend()

plt.savefig("Survival_rate.png")
                
plt.show()
```

```python
#calculate percentage changes in tumor size for each drug

#re-orient pivot table to have drug as index and time as columns
volume_change_df = round(tumors_df.pivot(index="Drug", columns = "Timepoint", values= "Tumor Volume (mm3)" ),2)

#create list of drug names by creating array from index values and then converting to a list
drug_array=volume_change_df.index.values
drug_list = drug_array.tolist()

#create new colume for percentage of tumor volume change
volume_change_df["Tumor Change"]=round((volume_change_df.iloc[:,-1]-volume_change_df.iloc[:,0])*100/volume_change_df.iloc[:,0],2)

#check values for percent changes are correct
volume_change_df["Tumor Change"]
```

```python
#create new column for conditional on whether drug is passing or not based on whether tumor grows or shrinks over trial
volume_change_df["Passing"] = np.where(volume_change_df["Tumor Change"]>=0, "no", "yes")

volume_summary_df = volume_change_df[["Tumor Change", "Passing"]]
volume_summary_df = volume_summary_df.loc[["Capomulin", "Infubinol", "Ketapril","Placebo"]]
volume_summary_df
```

```python
#plot for tumor change over 45 days of treatment

#x axis is index values of volume summary
x_axis=volume_summary_df.index.values

#y axis values are rounded percentage values of tumor change column
y_axis=round(volume_summary_df["Tumor Change"],1)

#conditional splits on y_axis values for <0 "passing" and >0 "failing"
mask1 = y_axis < 0
mask2 = y_axis >= 0

#passing values
plt.bar(x_axis[mask1], y_axis[mask1], color = 'green', width=-1, align="edge")

#failing values
plt.bar(x_axis[mask2], y_axis[mask2], color = 'red', width = -1, align="edge")

#set y ticks to be twenty units apart
plt.yticks(np.arange(-20, 61, 20))
plt.ylim(-30, 70)

#add grid
plt.grid()

#add title and y-label
plt.title("Tumor Change Over 45 Days Treatment")
plt.ylabel("% Tumor Volume Change")

#add text for bars
plt.text(-0.5, -7, str(f"{y_axis[0]}%"), ha='center', va='bottom', color="white")
plt.text(0.5, 0, str(f"{y_axis[1]}%"), ha='center', va='bottom', color="white")
plt.text(1.5, 0, str(f"{y_axis[2]}%"), ha='center', va='bottom', color="white")
plt.text(2.5, 0, str(f"{y_axis[3]}%"), ha='center', va='bottom', color="white")

plt.savefig("Tumor_Change_45days.png")
        
plt.show()
```
